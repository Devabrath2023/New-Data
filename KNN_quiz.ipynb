```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df1["status_id"]=le.fit_transform(df1["status_id"])
```

```python
df1["status_type"]=le.fit_transform(df1["status_type"])
```

```python
df1["status_published"]=le.fit_transform(df1["status_published"])
```

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_new=sc.fit_transform(df1)
```

```python
df_new
```

```python
df_scaled=pd.DataFrame(df_new,columns=["status_id","status_type","status_published","num_reactions","num_comments","num_shares",
                                       "num_likes","num_loves","num_wows","num_hahas","num_sads","num_angrys"])
```

```python
df_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
clusters=list(range(2,10))
ssd=[]
for n in clusters:
    model_clus=KMeans(n_clusters=n, max_iter=150,random_state=50)
    model_clus.fit(df_scaled)
    ssd.append(model_clus.inertia_)
```

```python
ssd
```

```python
plt.plot(clusters,ssd)
```

```python
Kmodel = KMeans(n_clusters =5, max_iter=150,random_state= 50)
Kmodel.fit(df_scaled)
```

```python
df1["ClusterId"]=Kmodel.labels_
```
